```python
import os
import pandas as pd 
import cv2  
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from sklearn.manifold import TSNE
import scipy
from scipy.misc import imread, imresize
```

```python
train_path = '../data/samples10-m'
#train_path = '../data/train'
cameras = os.listdir(train_path)

train_images = []
for camera in cameras:
    for fname in sorted(os.listdir(train_path+ '/'+ camera)):
        train_images.append((camera, fname))

train = pd.DataFrame(train_images, columns=['camera', 'fname'])
print(train.shape)
train.sample(5)
```

```python
images = []
cameras = []
input_path = '../data/samples10-m/'
#input_path = '../data/train/'

for img in train_images:
    im_path = input_path+img[0]+'/'+img[1]
    image = cv2.imread(im_path)    
    ni = image.reshape(-1)
    images.append (ni)
    cameras.append(img[0])

print('Images shape:', np.array(images).shape)
```

```python
y = np.array(cameras)
y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(images, cameras, test_size=0.2)
```

```python
from sklearn import svm
clf = svm.SVC()
clf.fit(X_train, y_train) 
```
